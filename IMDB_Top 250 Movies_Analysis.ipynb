```python
def convert_count_of_ratings(value):
    if 'M'in value:
        return float(value.replace('M','')) * 1000000
    elif 'K' in value:
        return float(value.replace('K',''))*1000
    else:
        return float(value)
```

**Understanding Domestic Gross and Domestic Weekend Gross Ratio | Revenue generation standpoint**

By dividing the weekend gross by the total domestic gross, you get a ratio that helps understand the significance of the opening weekend in contributing to the overall domestic earnings of the movie. 

A higher ratio indicates that a larger portion of the movie's earnings came from its opening weekend, highlighting the importance of a strong opening.

The opening weekend gross ratio is a key measure in understanding how much of a movie's earnings come from its initial release. 

A higher ratio suggests that a significant portion of the movie's total earnings were generated during its opening weekend. 

This underscores the importance of a strong opening, as it can greatly influence the **overall success of the movie**.

```python
domestic_weekend_gross_ratio = df1.sort_values(by= 'Domestic Weekend Gross Ratio (%)', ascending= False)
domestic_weekend_gross_ratio.head(10).round(2)
```
